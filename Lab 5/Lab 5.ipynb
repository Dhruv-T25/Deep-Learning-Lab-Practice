## For step functions
### weights are
* OR -> w[1,1], B[-0.5]
* NAND -> w[-1,-1], B[1.5]
* AND -> W[1,1], B[-1.5]

```python
def step_function(array):
  return (array >= 0).astype(int)
```

```python
import numpy as np

def step_function(array):
    return (array >= 0).astype(int)

inputs = np.array([[0,0],
                   [0,1],
                   [1,0],
                   [1,1]])

# OR and NAND weights
W1 = np.array([[1, -1],
               [1, -1]])

B1 = np.array([-0.5, 1.5])

l1 = step_function(inputs @ W1 + B1)

# AND weights
W2 = np.array([[1],
               [1]])

B2 = np.array([-1.5])

l2 = step_function(l1 @ W2 + B2)

print("Output of MLP :",l2.tolist())
```

## Sigmoid step function

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Data (XOR)
inputs = np.array([[0,0],
                   [0,1],
                   [1,0],
                   [1,1]])

targets = np.array([[0],
                    [1],
                    [1],
                    [0]])

# Initialize weights
W1 = np.random.randn(2,2)
B1 = np.zeros((1,2))

W2 = np.random.randn(2,1)
B2 = np.zeros((1,1))

lr = 0.1
epochs = 1000
```

```python
for i in range(epochs):

    # -------- Forward Pass --------
    z1 = inputs @ W1 + B1
    a1 = sigmoid(z1)

    z2 = a1 @ W2 + B2
    output = sigmoid(z2)


    error = targets - output  # Error rate

    # -------- Backprop --------
    delta2 = (output - targets) * sigmoid_derivative(output)

    dW2 = a1.T @ delta2
    dB2 = np.sum(delta2, axis=0, keepdims=True)

    delta1 = (delta2 @ W2.T) * sigmoid_derivative(a1)

    dW1 = inputs.T @ delta1
    dB1 = np.sum(delta1, axis=0, keepdims=True)


    # -------- Gradient Update --------
    W2 -= lr * dW2
    B2 -= lr * dB2
    W1 -= lr * dW1
    B1 -= lr * dB1

    if i % 100 == 0:
        loss = np.mean(error**2)
        print(f"Epoch {i}, Loss: {loss}")
print(f"Epoch {i}, Loss: {loss}")
```
