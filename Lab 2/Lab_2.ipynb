## Q3. Add logic to stop the training when the error is ‘zero’ for all the inputs.(Early stoping)

````python
import numpy as np

x_list = np.array([[0,0],[0,1],[1,0],[1,1]])
y_list = np.array([0,0,0,1])
epochs = 10
w,b = np.random.randint(1, 3, size=2),np.random.randint(1,3, size=1)
main_epo = 0

for _ in range(epochs):
  total_error_this_epoch = 0 # Initialize total error for the current epoch
  epoch_errors = [] # To store errors for printing after the epoch

  for x,y in zip(x_list,y_list):
    main_epo += 1
    z = np.dot(x,w) + b

    y_pred = 1 if z >= 0 else 0
    E = y - y_pred
    total_error_this_epoch += abs(E) # Accumulate absolute error
    epoch_errors.append(f"error : {E} <- at epoch {main_epo}") # Store for printing

    # Only update weights and bias if there's an error
    if E != 0:
      w += E*x
      b += E

  # Print all errors and final weights/bias for the epoch after the inner loop
  for err_msg in epoch_errors:
    print(err_msg)
  print(w,b)
  print()

  # Early stopping condition: if total_error_this_epoch is 0, all inputs were classified correctly
  if total_error_this_epoch == 0:
    print("Early stopping: All errors are zero for this epoch.")
    break
````
